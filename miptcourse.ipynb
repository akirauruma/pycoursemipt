```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
#hist 
n, bins,  patches = plt.hist(x, 50, density = True, facecolor = 'g', alpha = 0.75) 

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.030, r'$\mu = 100, \ \sigma = 15$')
plt.text(50, 0.033, r'$\varphi_{\mu, \sigma^2}(x) = \frac{1}{\sigma\sqrt{2\pi}} \ e^{-\frac{(x - \mu)^2}{2\sigma^2}} = \frac{1}{\sigma} \varphi\left(\frac{x - \mu}{\sigma}\right), \quad x\in\mathbb{R}$', fontsize = 20, color = 'red')
plt.axis([40, 160, 0, 0.04])
plt.grid(True)
plt.show()
```

```python
x = y = np.arange(-10, 10, 0.1)
X, Y = np.meshgrid(x, y)

Z = np.zeros_like(X)
m = (X != 0) & (Y != 0)
Z[m] = np.sin(X[m]) * np.sin(Y[m]) / (X[m] * Y[m])


ax = plt.figure().add_subplot(111, projection="3d")
figure = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, cmap='turbo', lw = 0)
plt.colorbar(figure, shrink=0.6, aspect=12)
plt.show()

plt.savefig(r'/home/sutemou/code/mlka/pycoursemipt/figs/fig2')
```
