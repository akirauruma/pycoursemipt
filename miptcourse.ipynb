```python
x = np.arange(-10, 10.01, 0.01)
phi = np.arange(0, 2*np.pi, 0.01)
rho = 2 * phi
plt.plot(x, x**2)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), label=r'$f_1 = \sin(x)$')
plt.plot(x, np.cos(x), label=r'$f_2 = \cos(x)$')
plt.plot(x, x, label=r'$f_3 = x$')
plt.xlabel(r'$x$', fontsize = 14)
plt.ylabel(r'$f(x)$', fontsize = 14)
plt.grid(True)
plt.legend(loc = 'best', fontsize = 10)
plt.savefig(r'/home/sutemou/code/mlka/pycoursemipt/figs/fig0')
plt.show()
```

```python
#subplot 0
sp = plt.subplot(221)

plt.plot(x, np.sin(x))
plt.title(r'$f_0 = \sin(x)$')
plt.grid(True)

#subplot 1
sp = plt.subplot(222)

plt.plot(x, np.cos(x), 'g')
plt.axis('equal')
plt.title(r'$f_0 = \cos(x)$')
plt.grid(True)

#subplot 2
sp = plt.subplot(223)
t = np.arange(-10, 11, 1)

plt.plot(x, x** 2, t, t**2, 'ro')
plt.title(r'$f_0 = x^2$')

#subplot 3
sp = plt.subplot(224)

plt.plot(x, np.log(x), 'purple')
plt.title(r'$f_0 = \log(x)$')
plt.grid(True)


plt.savefig(r'/home/sutemou/code/mlka/pycoursemipt/figs/fig1')
```
